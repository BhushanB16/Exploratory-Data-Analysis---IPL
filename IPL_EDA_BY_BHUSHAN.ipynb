# **Data visualization and Storytelling:** 

```python
#Barplot of Top 5 Player of Match

plt.figure(figsize=(8, 5))
top_5_players = ipl['player_of_match'].value_counts().head(5)
plt.bar(top_5_players.index, top_5_players.values)
plt.xlabel('Player')
plt.ylabel('Number of Player of the Match Awards')
plt.title('Top 5 Players of the Match')
plt.xticks(rotation=45)
plt.show()
```

```python
#Getting the frequency pf result columns
ipl['result'].value_counts()
```

```python
#Find out No of Toss Wins w.r.t. Each Team
ipl['toss_winner'].value_counts()
```

```python
#Extracting the records where a team won batting first
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
#creating plot of Win_by_runs
plt.figure(figsize=(8, 5))
plt.hist(batting_first['win_by_runs'], bins=20, edgecolor='black')
plt.xlabel('Margin of Victory (Runs)')
plt.ylabel('Frequency')
plt.title('Distribution of Runs in Victories')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#No of WIns w.r.t. Rach Team After Batting First
batting_first['winner'].value_counts()
```

```python
#Creating barplot for Top 3 Teamsn Wins After Batting First
plt.figure(figsize=(6, 5))
top_3_teams = batting_first['winner'].value_counts().head(3)
plt.bar(top_3_teams.index, top_3_teams.values, color=['blue', 'orange', 'green'])
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.title('Top 3 Teams with Most Wins after Batting First')
plt.show()
```

```python
#Making Pie Chart

plt.figure(figsize=(7, 7))
labels = list(batting_first['winner'].value_counts().keys())
sizes = list(batting_first['winner'].value_counts())
plt.pie(sizes, labels=labels, autopct='%0.1f%%')
plt.xlabel('Teams')
plt.title('Distribution of Wins after Batting First')
plt.show()
```

```python
#Extracting those Records Where A Team has Won After Batting Second
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
batting_second.head()
```

```python
#Making Histrogram for Frequency of wins w.r.t. No. of Wickets
plt.figure(figsize=(7, 7))
plt.hist(batting_second['win_by_wickets'], bins=30, edgecolor='black')
plt.xlabel('Number of Wickets')
plt.ylabel('Frequency')
plt.title('Distribution of Wins by Wickets')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#finding out the frequency of no of wins w.r.t. each time after batting second
batting_second['winner'].value_counts()
```

```python
#Making a bar plot for Top 3 teamsn with wins after batting second

plt.figure(figsize=(7,7))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=['blue','green','orange'])
plt.show()
```

```python
#Making Pie chart for wins after batting second
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.xlabel('Teams')
plt.show()
```
